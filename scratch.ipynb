```python
import matplotlib.pyplot as plt
import numpy as np

# Confusion matrix values
A = 0  # True Positive
B = 0  # False Positive
C = 2  # False Negative
D = 9  # True Negative

confusion_matrix = np.array([[D, B], [C, A]])

fig, ax = plt.subplots()
cax = ax.matshow(confusion_matrix, cmap=plt.cm.Blues)

plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Confusion Matrix')
plt.xticks([0, 1], ['Actual Churned (1)', 'Actual Not Churned (0)'])
plt.yticks([0, 1], ['Predicted Churned (1)', 'Predicted Not Churned (0)'])
plt.colorbar(cax)

plt.show()
```

```python
# Confusion matrix values
A = 0  # True Positive
B = 0  # False Positive
C = 2  # False Negative
D = 9  # True Negative

# Calculate metrics
precision = A / (A + B) if (A + B) != 0 else 0
npv = D / (D + C) if (D + C) != 0 else 0
accuracy = (A + D) / (A + B + C + D)
specificity = D / (D + B) if (D + B) != 0 else 0
recall = A / (A + C) if (A + C) != 0 else 0

# Print the calculated metrics
print(f"A (True Positive): {A}")
print(f"B (False Positive): {B}")
print(f"C (False Negative): {C}")
print(f"D (True Negative): {D}")
print(f"Precision: {precision:.4f}")
print(f"NPV (Negative Predictive Value): {npv:.4f}")
print(f"Accuracy: {accuracy:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"Recall (Sensitivity): {recall:.4f}")
```
